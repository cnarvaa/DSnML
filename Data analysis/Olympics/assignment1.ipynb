* [45-55] and [15-25) correspondingly

## This is a way of creating categories from numerical data. See the second aggregation.

```python
import numpy as np
```

```python
age_cats_2008 = df[df["Year"]==2008].drop_duplicates("ID")
print(len(age_cats_2008.query("Age>=15 & Age <25")))
print(len(age_cats_2008.query("Age>=25 & Age <35")))
print(len(age_cats_2008.query("Age>=35 & Age <45")))
print(len(age_cats_2008.query("Age>=45 & Age <=55")))
```

```python
def age_category(age):
    '''Maps age to four categories'''

    if 15 <= age < 25:
        return '[15, 25)'
    elif 25 <= age < 35:
        return '[25, 35)'
    elif 35 <= age < 45:
        return '[35, 45)'
    return '[45, 55]'


# map() applies age_category() function to every value in data.Age
age_cats_2008['age_category'] = age_cats_2008.Age.map(age_category)
(age_cats_2008 
.groupby('age_category')
.size())
```
